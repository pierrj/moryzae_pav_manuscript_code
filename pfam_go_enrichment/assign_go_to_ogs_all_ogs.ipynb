```python
## import Orthogroups without 70-15
og_no_70_15 = '../../pipeline_methods/Orthogroups.tsv'
```

```python
og_no_70_15_df = pd.read_csv(og_no_70_15, dtype='string', sep='\t', index_col = 0)
```

```python
# used to get the og a gene belongs to

df_lol = og_no_70_15_df.values.tolist()

og_dict_no_70_15 = {}

for i, row in enumerate(df_lol):
    og = og_no_70_15_df.index[i]
    for cell in row:
        if not pd.isnull(cell):
            for protein in cell.split(', '):
                og_dict_no_70_15[protein] = og
```

```python
# use to get the group of genes assigned to an og

genes_per_og_no_70_15 = {}

for gene in og_dict_no_70_15:
    og = og_dict_no_70_15[gene]
    if og not in genes_per_og_no_70_15:
        genes_per_og_no_70_15[og] = []
    genes_per_og_no_70_15[og].append(gene)
```

```python
PPV_cutoff = 0.6
```

```python
go_terms_per_og_dict = {}

column_names = ["qpid", "ontology", "goid", "desc", "ARGOT_score", "ARGOT_PPV", "ARGOT_rank", "goclasscount"]


for og in genes_per_og_no_70_15:
    go_terms_per_og_dict[og] = pd.DataFrame(columns = column_names)
```

```python
for proteome in proteomes:
    print(proteome)
    kparse_out = "all_go/GO."+ proteome + ".out"
    df = pd.read_csv(kparse_out, sep = '\t')
    if len(df.index) < 250000:
        print(len(df.index))
        raise ValueError
    ## filter to ppv cutoff
    df = df[df.ARGOT_PPV >= PPV_cutoff]
    ## only keep the first ranked term for each
    df = df[df.ARGOT_rank == 1]
    for og in genes_per_og_no_70_15:
        genes_in_og = genes_per_og_no_70_15[og]
        df_subset = df[df['qpid'].isin(genes_in_og)]
        go_terms_per_og_dict[og] = pd.concat([go_terms_per_og_dict[og], df_subset], ignore_index=True, sort=False)
```

```python
with open('Orthogroups_GO_ids.txt', 'w', newline = '') as output_csv:
    w = csv.writer(output_csv, delimiter = '\t')
    for og in go_terms_per_og_dict:
        go_df = go_terms_per_og_dict[og]
        go_ids_per_og = []
        for go in go_df.goid:
            if str(go).zfill(7) not in go_ids_per_og:
                go_ids_per_og.append(str(go).zfill(7))
        if len(go_ids_per_og) == 0:
            continue
        elif len(go_ids_per_og) == 1:
            ids = 'GO:' + go_ids_per_og[0]
        else:
            ids = []
            for go_id in go_ids_per_og:
                ids.append('GO:'+go_id+', ')
            ids = ''.join(ids)
            ids = ids[:-2]
            print(ids)
        w.writerow([og, ids])
```
