```python
import pandas as pd
import csv
from collections import Counter
```

```python
orthogroups = 'Orthogroups.tsv'
```

```python
orthogroups_df = pd.read_csv(orthogroups, dtype='string', sep='\t', index_col = 0)
```

```python
df_lol = orthogroups_df.values.tolist()
```

```python
og_dict = {}

for i, row in enumerate(df_lol):
    og = orthogroups_df.index[i]
    for cell in row:
        if not pd.isnull(cell):
            for protein in cell.split(', '):
                og_dict[protein] = og
```

```python
counts_per_og = {}

for i, row in enumerate(df_lol):
    og = orthogroups_df.index[i]
    proteins_in_og = []
    for cell in row:
        if not pd.isnull(cell):
            for protein in cell.split(', '):
                proteins_in_og.append(protein)
    counts_per_og[og] = len(proteins_in_og)
```

```python
effectors_file = 'all_effector_names'

with open(effectors_file, newline = '') as file:
    file_reader = csv.reader(file, delimiter = '\t')
    effectors = [ row[0] for row in file_reader]
```

```python
orthogroups = []
not_in_orthogroups = []
for effector in effectors:
    try:
        og = og_dict[effector]
    except KeyError:
        print('{} not in orthogroups for some reason'.format(effector))
        not_in_orthogroups.append(effector)
        continue
# if og not in orthogroups:
    orthogroups.append(og)
```

```python
print(len(orthogroups))
print(len(not_in_orthogroups))
```

```python
# counts
c = Counter(orthogroups)

orthogroups_more_than_half_effectors = []

for og in c:
    if c[og] > counts_per_og[og]:
        raise ValueError
    print(og)
    print(c[og]/counts_per_og[og])
    print(c[og])
    print(counts_per_og[og])
    if c[og]/counts_per_og[og] > 0.5:
        orthogroups_more_than_half_effectors.append(og)
```
